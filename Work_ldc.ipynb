```python
import pandas as pd

# Assuming you have a DataFrame with a column named 'json_data' containing JSON strings
# Load your DataFrame
df = Credit_notes  # Load your data from a CSV file, adjust the file name as needed

# Function to extract buyers' names from JSON data
def extract_buyers_name(json_str):
    try:
        parsed_json = json.loads(json_str)
        return parsed_json['buyersName']
    except (KeyError, json.JSONDecodeError):
        return None

# Apply the function to the 'json_data' column to create a new column 'buyersName'
df['buyersName'] = df['jsonData'].apply(extract_buyers_name)

# Print the DataFrame to see the result
df
```
